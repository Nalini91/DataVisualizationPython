### Step-1: Importing Packages

```python
#import all librarys and modules needed
import pandas as pd
import numpy as np
import seaborn as sb
import matplotlib.pyplot as plt
from matplotlib import style
style.use('classic')
plt.rcParams['figure.figsize'] = (20,10)
```

### Step-2 : Importing and Cleaning Data 

```python
# #import and clean data, remember path is data/Canada.xlsx

immigration_df= pd.read_excel('C:/Users/Buddi/Documents/GitHub/DataVisualizationPython/Data/Canada.xlsx',1, skiprows = range(20), skipfooter = 2)
```

```python
immigration_df.drop(['AREA','REG','DEV','Type','Coverage','DevName'], axis=1, inplace=True)
```

```python
immigration_df.rename(columns = {'OdName':'country','AreaName':'continent','RegName':'region'}, inplace = True)
```

```python
immigration_df.duplicated().sum()
```

```python
immigration_df['total'] = immigration_df.sum(axis = 1)
```

```python
immigration_df = immigration_df.set_index('country')
```

```python
immigration_df.head()
```

```python
immigration_df.rename(index = {'United Kingdom of Great Britain and Northern Ireland':'UK & Ireland',
                               'United Arab Emirates':'UAE'}, inplace = True)
```

```python
immigration_df.columns = immigration_df.columns.astype(str)
```

```python
years = list(map(str, range(1980,2013)))
```

```python
immigration_df.head()
```

```python
immigration_df.dtypes
```

```python
immigration_df.columns
```

### Step-3 : Creating Beautiful Visualizations

#### Line Chart

```python
# Single line chart

fig1 = immigration_df.loc['Haiti', years].plot(kind = 'line', color = 'r')
plt.title('Immigration from Haiti to Canada from 1980-2013',color = 'black')
plt.xlabel('Years',color = 'black')
plt.ylabel('Number of Immigrants',color = 'black')
plt.xticks(color = 'black')
plt.yticks(color = 'black')
plt.savefig('linechart_single.png')

plt.show()

```

```python
# Multiple Line chart 

fig2 = plt.plot(immigration_df.loc['India',years], label = 'India')
plt.plot(immigration_df.loc['China',years], label = 'China')
plt.plot(immigration_df.loc['Philippines',years], label = 'Philippines')
# plt.legend(loc = 'upper left', fontsize = 12)
plt.legend(loc = 'lower right', fontsize = 12)
plt.xticks(rotation = 120, color = 'black')
plt.yticks(color = 'blue')
plt.title('Immigration to Canada from 1980-2013',color = 'black')
plt.xlabel('Year',color = 'black')
plt.ylabel('Number of Immigrants',color = 'black')
plt.savefig('linechart_multiple.png')

plt.show()
```

#### Bar Chart

```python
# Vertical bar chart

# style.use('ggplot')

immigration_df_tot = pd.DataFrame(immigration_df.loc[:,years].sum())
immigration_df_tot.rename(columns = {0:'total'}, inplace = True)

immigration_df_tot.plot(kind = 'bar', legend = True)
plt.title('Total Immigrants to Canada from 1980-2013',color = 'white')
plt.xticks(color = 'white')
plt.yticks(color = 'white')
plt.xlabel('Year',color = 'white')
plt.ylabel('Number of Immigrants',color = 'white')
plt.savefig('bar_vertical.png')

plt.show()
# Do not change the style back to ggplot 
# delete the style.use('ggplot') line of code



```

Notice that style is still set to cyberpunk.  How do we fix it so we can see the labels?<br> <br>

Answer: change the color = 'black' to 'white'

Feel free to experiment and add any notes or observations here. <br><br><br><br><br>

By setting legend to True we are able to see the 'Total' title on the top right.

```python
# Horizontal bar chart
#change style back to ggplt

style.use('ggplot')
immigration_df_top10 = pd.DataFrame(immigration_df.nlargest(10,'total')['total'].sort_values(ascending = False))

immigration_df_top10.plot.barh(legend = True, color = 'blue', edgecolor = 'black')
plt.title('Top 10 Immigrant Countries to Canada from 1980-2013',color = 'black')
plt.xlabel('Number of Immigrants',color = 'black')
plt.ylabel('Country',color = 'black')
plt.xticks(color = 'black')
plt.yticks(color = 'black')
plt.savefig('bar_horizontal.png')

plt.show()
```

Feel free to experiment and add any notes or observations here. <br><br><br><br><br>

changed sort values to sort_values(ascending = False) so the bar graphs are displayed as country having lowest number of immigrants first.


```python
# Grouped bar chart
year_int10 = list(map(str, (1980,1990,2000,2010, 2013)))
immigration_df_group = pd.DataFrame(immigration_df.loc[['India','China','Philippines','Pakistan'],year_int10].T)

immigration_df_group.plot.bar(edgecolor = 'white')
plt.title('Total Immigrants to Canada from 1980-2013',color = 'black')
plt.xticks(color = 'black')
plt.yticks(color = 'black')
plt.xlabel('Year',color = 'black')
plt.ylabel('Number of Immigrants',color = 'black')
plt.legend(title = 'Country', fontsize = 12, labelcolor='k')
plt.savefig('bar_grouped.png')

plt.show()
```

Notice how the labels in the legend have disapeared?  We can fix this by adding labelcolor='k' to plt.legend<br>
<br>plt.legend(title = 'Country', fontsize = 12, labelcolor='black')

#### Area Chart

```python
# Area Chart

plt.fill_between(immigration_df_tot.index, immigration_df_tot['total'], color="gray", alpha=0.4)
plt.plot(immigration_df_tot.index, immigration_df_tot['total'], color = 'Slateblue', alpha = 0.6)
plt.title('Total Immigrants to Canada from 1980-2013', fontsize = 18, color = 'black')
plt.yticks(fontsize = 12, color = 'black')
plt.xticks(fontsize = 12, rotation = 90, color = 'black')
plt.xlabel('Year', fontsize = 14, color = 'black')
plt.ylabel('Number of Immigrants', fontsize = 14, color = 'black')
plt.savefig('area_simple.png')

plt.show()
```

Feel free to experiment and add any notes or observations here. <br><br><br><br><br>

I have changed the fill_between color to gray.Also,the axis lines disappear when the alpha value is set to 1.The lines were clear when the value <1.As the value increases the axis lines disappears and is difficult to analyze the results.

```python
# cyberpunk simple area chart


import mplcyberpunk

style.use('cyberpunk')

plt.fill_between(immigration_df_tot.index, immigration_df_tot['total'], color = 'greenyellow', alpha = 0.1)
plt.plot(immigration_df_tot.index, immigration_df_tot['total'], color = 'palegreen', alpha = 1)
mplcyberpunk.add_glow_effects()
plt.title('Total Immigrants to Canada from 1980-2013', fontsize = 20,fontweight = 'bold', color = 'white')
plt.yticks(fontsize = 14, color = 'white',fontweight = 'bold')
plt.xticks(fontsize = 14, rotation = 90, color = 'white',fontweight = 'bold')
plt.xlabel('Year', fontsize = 16, color = 'white',fontweight = 'bold')
plt.ylabel('Number of Immigrants', fontsize = 16, color = 'white',fontweight = 'bold')
plt.savefig('cyber_area_simple.png')

plt.show()

```

Feel free to experiment and add any notes or observations here. <br><br><br><br><br>

add_glow_effects() makes the area chart and outline more glowy.Setting font_weight to 'Bold' makes the lables,titles in bold format.

```python
# stacked area chart

color = ['skyblue','lightcoral']
top2_list = immigration_df.nlargest(2, 'total').index.tolist()
immigration_df_top2 = pd.DataFrame(immigration_df.loc[top2_list, years].T)

immigration_df_top2.plot(kind = 'area', stacked = True, color = color)
plt.title('Total Immigrants to Canada from 1980-2013',color = 'black')
plt.legend(fontsize = 12)
plt.xlabel('Year',color = 'black')
plt.ylabel('Number of Immigrants',color = 'black')
plt.xticks(color = 'black')
plt.yticks(color = 'black')
plt.savefig('area_stacked.png')

plt.show()
```

Feel free to experiment and add any notes or observations here. <br><br><br><br><br>

If 'stacked = True' the charts are not interfering with each other and clearly stacked ,while if 'stacked = False' the area charts are not clear and merge with each other which makes analysis difficult.

```python
# unstacked area chart

immigration_df_top2.plot(kind = 'area', stacked = False, color = color)
plt.title('Total Immigrants to Canada from 1980-2013',color = 'black')
plt.xlabel('Year',color = 'black')
plt.ylabel('Number of Immigrants',color = 'black')
plt.legend(fontsize = 12)
plt.xticks(color = 'black')
plt.yticks(color = 'black')
plt.savefig('area_unstacked.png')

plt.show()
```

Feel free to experiment and add any notes or observations here. <br><br><br><br><br>

If 'stacked = True' the charts are not interfering with each other and clearly stacked ,while if 'stacked = False' the area charts are not clear and merge with each other which makes analysis difficult.

#### Scatter Plot

With the newist version of Seaborn we have to specify x and y.

example: sb.scatterplot(x = 'sepal_length', y = 'sepal_width', data = df_iris)

```python
#scatter plot comparing sepal length to sepal width

import seaborn as sb
immigration_df_iris = sb.load_dataset('iris')

sb.scatterplot('sepal_length','sepal_width', data = immigration_df_iris, s = 200, linewidth = 3, edgecolor = 'Red')
plt.title('Scatter Plot sample', color = 'black', fontsize = 18)
plt.savefig('scatter.png')

plt.show()
```

Feel free to experiment and add any notes or observations here. <br><br><br><br><br>

Changing 's' value increases/decreases the center of each scatter plot and changing linewidth increases/decreases the outer line of scatterplot.

#### Histogram

With the newist version of Seaborn they have deprecated distplot. Replace distplot with histplot and add kde = True

example:
sb.histplot(df_iris['sepal_length'], color = 'Red', label = 'Sepal Length', kde = True)

```python
#Histogram side by side, with kde

df_iris = sb.load_dataset('iris')

sb.histplot(df_iris['sepal_length'], color = 'Red', label = 'Sepal Length',kde = True)
sb.histplot(df_iris['sepal_width'], color = 'skyblue', label = 'Sepal Width',kde = True)
plt.legend(fontsize = 12)
plt.xlabel('Sepal Width', color = 'black')
plt.ylabel('Frequency', color = 'black')
plt.xticks(color = 'black')
plt.yticks(color = 'black')
plt.savefig('histogram.png')

plt.show()
```

Feel free to experiment and add any notes or observations here. <br><br><br><br><br>

when given kde = True a line is displayed on histogram which represents the frequency/no.of occurences of a single variable i.e, sepal_width values.This helps to clearly see the results.

#### Doughnut Chart

```python
# Doughnut Chart


top5_list = immigration_df.nlargest(5, 'total').index.tolist()
immigration_df_top5 = pd.DataFrame(immigration_df.loc[top5_list, 'total'].T)
circle = plt.Circle( (0,0), 0.7, color='purple')
colors = ['gold', 'yellowgreen', 'lightcoral', 'lightskyblue', 'lightgreen', 'pink']

plt.pie(immigration_df_top5['total'], autopct = '%1.1f%%', shadow = True, explode = [0.1,0,0,0,0], colors = colors, startangle = 90)
fig = plt.gcf()
fig.gca().add_artist(circle)
plt.legend(immigration_df_top5.index, fontsize = 12, loc = 'upper left')
plt.title('Top 5 Immigrant Country Distribution', color = 'black', fontsize = 18)
plt.axis('equal')
plt.savefig('doughnut.png')

plt.show()
```

#### Regression Plot

```python
# Strong trend

immigration_df_reg = pd.DataFrame(immigration_df.loc['China',years])
immigration_df_reg.reset_index(inplace = True)
immigration_df_reg.rename(columns = {'index':'year'}, inplace = True)
immigration_df_reg[['year','China']] = immigration_df_reg[['year','China']].astype(int)

sb.regplot(x = 'year', y = 'China', data = immigration_df_reg, color = 'b', scatter_kws = {'s':150,'alpha':0.4})
plt.title('Total Immigration from China to Canada from 1980-2013', color = 'black')
plt.xlabel('Year', color = 'black')
plt.ylabel('Number of Immigrants', color = 'black')
plt.xticks(color = 'black')
plt.yticks(color = 'black')
plt.savefig('reg_strong.png')

plt.show()
```

Feel free to experiment and add any notes or observations here. <br><br><br><br><br>

```python
# Weak trend

immigration_df_reg1 = pd.DataFrame(immigration_df.loc[['Denmark','Norway','Sweden'],years].sum())
immigration_df_reg1.reset_index(inplace = True)
immigration_df_reg1.rename(columns = {'index':'year',0:'total'}, inplace = True)
immigration_df_reg1[['year','total']] = immigration_df_reg1[['year','total']].astype(int)

sb.regplot(x = 'year', y = 'total', data = immigration_df_reg1, color = 'Red', scatter_kws = {'s':150,'alpha':0.4})
plt.title('Total Immigrants from Scandinavia to Canada from 1980-2013', color = 'black')
plt.xticks(color = 'black')
plt.yticks(color = 'black')
plt.xlabel('Year', color = 'black')
plt.ylabel('Number of Immigrants', color = 'black')
plt.savefig('reg_weak.png')

plt.show()
```
